```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

## 1 Data Exploration

### 1.1 Load Data

```python
train_path = 'data/train.jsonl'
train_file = open(train_path,encoding='utf-8')
train_data = [json.loads(line) for line in train_file]
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
train_data = pd.json_normalize(train_data)
```

### 1.2 Drop Columns

```python
remain_list = ['string','label','label_confidence']
```

```python
train = train_data[remain_list]
# test = test_data[remain_list]
```

```python
train
```

### 1.3 Check Data

Check the Null value

```python
train["string"].isnull().any()
```

```python
train["label"].isnull().any()
```

Check the duplicate

```python
train.drop_duplicates(['string'])
```

### 1.4 Data Distribution

```python
plt.figure(figsize=(14,6),dpi=80)
plt.subplot(1,3,1)
plt.hist([len(s.split()) for s in train["string"]])
plt.xlabel('Number of words in a sample')
plt.ylabel('Number of samples')
plt.title('Sample length distribution')

plt.subplot(1,3,2)
plt.hist(train["label"])
plt.xlabel('Label of classes')
plt.ylabel('Number of samples')
plt.title('Class label distribution')

plt.subplot(1,3,3)
plt.hist(train["label_confidence"])
plt.xlabel('Label of classes')
plt.ylabel('Number of samples')
plt.title('Label confidence distribution')

plt.show()
```
